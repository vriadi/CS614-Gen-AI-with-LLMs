```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Input text
prompt = 'Hello, welcome to SMU MITB Gen AI and LLM Course. What is your name?'
```

```python
input_ids = tokenizer.encode(prompt, return_tensors="pt")
```

```python
output = model.generate(
    input_ids,
    max_length=50,
    num_return_sequences=1,
    temperature=1.0, # Set the temperature here
    do_sample=True, # Ensure sampling is enabled for temperature to have an effect
)
```

```python
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

```python
output2 = model.generate(
    input_ids,
    max_length=50,
    num_return_sequences=1,
    temperature=25.0, # Set the temperature here
    do_sample=True, # Ensure sampling is enabled for temperature to have an effect
)
```

```python
generated_text2 = tokenizer.decode(output2[0], skip_special_tokens=True)
print(generated_text2)
```
